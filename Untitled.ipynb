```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df1 = pd.read_csv('data/Accused-Witches-Data-Set.csv')
```

```python
df1.head()
df1.columns = ["accused_witch", "residence", "month_of_accusation", "month_of_execution", "sort"]
```

```python
df1["residence"]
```

```python
places = df1.groupby(["residence"])["residence"].count()
places = places.to_frame("places_count").reset_index()
places
```

```python
timeline_acc = df1.groupby(["month_of_accusation"])["month_of_accusation"].count()
timeline_acc = timeline_acc.to_frame("accusation_count").reset_index()
timeline_acc = timeline_acc.rename(columns={"month_of_accusation": "month"})
timeline_acc
```

```python
timeline_ex = df1.groupby(["month_of_execution"])["month_of_execution"].count()
timeline_ex = timeline_ex.to_frame("execution_count").reset_index()
timeline_ex = timeline_ex.rename(columns={"month_of_execution": "month"})
timeline_ex[["month"]] = timeline_ex[["month"]].astype(int)
timeline_ex
```

```python
timeline = timeline_acc.merge(timeline_ex, how="left", on="month")
timeline["execution_count"].fillna(0, inplace = True)
timeline[["execution_count"]] = timeline[["execution_count"]].astype(int)
timeline = timeline[timeline.month != -1]
timeline
```

```python
plt.plot(timeline['month'], timeline['accusation_count'])
plt.plot(timeline['month'], timeline['execution_count'])
plt.show()
```
